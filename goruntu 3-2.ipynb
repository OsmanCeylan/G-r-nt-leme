```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import imshow

img_1=plt.imread("img.jpg")
img_2=plt.imread("img.jpg")

%matplotlib inline

imshow(img_1)
```

```python
def inverse(pixel):
    return 255-pixel
    
```

```python
m = img_1.shape[0]
n = img_1.shape[1]
    
print (m,n)
    
for i in range (m):
    for j in range (n):
        img_2[i,j,0] = inverse(img_1[i,j,0])
        img_2[i,j,1] = inverse(img_1[i,j,1])
        img_2[i,j,2] = inverse(img_1[i,j,2])
imshow(img_2)
```
