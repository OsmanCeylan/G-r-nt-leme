```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import imshow

img_1=plt.imread("wall.jpg")

%matplotlib inline
imshow(img_1)
```

```python
def rotate_img_right(wimage):
    
    m=wimage.shape[0]
    n=wimage.shape[1]
    
    last= [[[0 for col in range(3)]for row in range(m)] for x in range(n)]
    
    for i in range (m):
        for j in range (n):
            last[j][m-i-1][0]=int(wimage[i,j,0])
            last[j][m-i-1][1]=int(wimage[i,j,1])
            last[j][m-i-1][2]=int(wimage[i,j,2])
            
    imshow(last)

def rotate_img_left(wimage):
    
    m=wimage.shape[0]
    n=wimage.shape[1]
    
    last= [[[0 for col in range(3)]for row in range(m)] for x in range(n)]
    
    for i in range (m):
        for j in range (n):
            last[n-j-1][i][0]=int(wimage[i,j,0])
            last[n-j-1][i][1]=int(wimage[i,j,1])
            last[n-j-1][i][2]=int(wimage[i,j,2])
            
    imshow(last)

def rotate_img_right_and_simetri(wimage):
    
    m=wimage.shape[0]
    n=wimage.shape[1]
    
    last= [[[0 for col in range(3)]for row in range(m)] for x in range(n)]
    
    for i in range (m):
        for j in range (n):
            last[n-j-1][m-i-1][0]=int(wimage[i,j,0])
            last[n-j-1][m-i-1][1]=int(wimage[i,j,1])
            last[n-j-1][m-i-1][2]=int(wimage[i,j,2])
            
    imshow(last)
```

```python
rotate_img_right(img_1)
```

```python
rotate_img_left(img_1)
```

```python
rotate_img_right_and_simetri(img_1)
```
